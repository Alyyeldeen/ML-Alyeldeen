# PREPROCESSING OF QUANTITIVE TRAINING DATASET

**1) IMPUTATION OF MISSING VALUES**

```python
# Calculate the mean of EHQ_EHQ_Total (excluding missing values)
mean_ehq = df_quant['EHQ_EHQ_Total'].mean()

# Replace missing values with the calculated mean
df_quant['EHQ_EHQ_Total'] = df_quant['EHQ_EHQ_Total'].fillna(mean_ehq)
```

**The missing values in EHQ_EHQ_Total were imputed with the mean score of 59.51 because our data shows strong right-handed dominance (mean=59.51, median=77.8), and research confirms approximately 90% of people are right-handed as I found when searching online so using the mean provides the most statistically valid estimate for these missing cases.**

```python
#replace with median 14
df_quant['ColorVision_CV_Score'] = df_quant['ColorVision_CV_Score'].fillna(14)
```

**Ok so i decided to replace the ColorVision_CV_Score missing values with the median score as we can observe from the .describe that the 25% quantile and the 75% quantile values are 14 and the max value is 14 so this means that from 25% to 100% of the people in the dataset scored 14 which is the value that assumes normal color vision therefore using 14 for missing values makes the safest assumption that those people probably had normal vision too rather than guessing they might have color blindness.**

```python
apq_cols = ['APQ_P_APQ_P_CP', 'APQ_P_APQ_P_ID', 'APQ_P_APQ_P_INV', 
            'APQ_P_APQ_P_OPD', 'APQ_P_APQ_P_PM', 'APQ_P_APQ_P_PP']
df_quant[apq_cols] = df_quant[apq_cols].fillna(df_quant[apq_cols].median())
```

**I used the median to fill missing values in all APQ parenting scores as i noticed the mean and median are so similar for these scales, so i chose the median as it works better when handling ordinal data (like survey ratings) properly and isn't affected by extreme scores. This gives us the most typical parenting behavior values for missing cases**

```python
# Define all SDQ features present in your data
sdq_features = [
    'SDQ_SDQ_Conduct_Problems',
    'SDQ_SDQ_Difficulties_Total',
    'SDQ_SDQ_Emotional_Problems',
    'SDQ_SDQ_Externalizing',
    'SDQ_SDQ_Generating_Impact',
    'SDQ_SDQ_Hyperactivity',
    'SDQ_SDQ_Internalizing',
    'SDQ_SDQ_Peer_Problems',
    'SDQ_SDQ_Prosocial'
]

# Impute missing values with median (recommended for clinical scales)
df_quant[sdq_features] = df_quant[sdq_features].fillna(df_quant[sdq_features].median())
```

**I used the median because SDQ scores are ordinal and the median represents typical responses better than the mean, avoiding distortion from extreme values. This ensures sensible imputations**

```python
#imputing with the median
df_quant['MRI_Track_Age_at_Scan'] = df_quant['MRI_Track_Age_at_Scan'].fillna(df_quant['MRI_Track_Age_at_Scan'].median())
```

**This method fills in missing MRI scan ages by using the median of the feature itself So we used the overall median age to ensure reliable estimates while preserving the natural relationship between parenting behaviors and scan timing.**

```python
# Check for missing values after imputing all of them
print(df_quant.isnull().sum())

# Check for duplicates
print(f"Number of duplicates: {df_quant.duplicated().sum()}")
```

```python
df_categ['PreInt_Demos_Fam_Child_Ethnicity'] = df_categ['PreInt_Demos_Fam_Child_Ethnicity'].fillna(3)  # 3="Unknown"
```

**Using the mode assumes missing values match the most common group which could introduce bias if certain ethnicities are more likely to have missing data so I chose to mark them as (3) which is = "Unknown" as it is already an existing category to keep the data honest and avoids false assumptions. This is especially important for sensitive demographics like ethnicity where incorrect imputation could skew results**

```python
df_categ['PreInt_Demos_Fam_Child_Race'] = df_categ['PreInt_Demos_Fam_Child_Race'].fillna(10)  # 10="Unknown"
```

**I used here the same concept as the imputation above as there is an already existing feature called unknown = 10 so I used it in imputation rather than the mode**

```python
# Calculate the mode of the column
mode_value = df_categ['MRI_Track_Scan_Location'].mode()[0]  # [0] to get the first mode if multiple exist

# Replace NaN values with the mode
df_categ['MRI_Track_Scan_Location'].fillna(mode_value, inplace=True)
```

**I decided to replace the missing MRI_Scan_location values with the mode of the feature itself as they only represent 3 entries of the whole dataset so this will not alter or effect the data neagtively or give bias to any category without hurting analysis**

```python
df_categ['Barratt_Barratt_P1_Edu'] = df_categ['Barratt_Barratt_P1_Edu'].fillna(df_categ['Barratt_Barratt_P1_Edu'].mode()[0])  
```

**The mode is ideal for imputing missing education data because it preserves the most common category without distorting the dataset's natural distribution. This approach is simple and avoids introducing unrealistic values**

```python
df_categ['Barratt_Barratt_P1_Occ'] = df_categ.groupby('Barratt_Barratt_P1_Edu')['Barratt_Barratt_P1_Occ'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)
```

**Grouping by education and imputing the mode for missing occupations makes sense because people with similar education levels often work in similar jobs as education and occupation strongly links so this method keeps the data realistic as college grads wont be assigned jobs typical for high school grads and vice versa so this avoids distorting the dataset**

```python
df_categ['Barratt_Barratt_P2_Edu'] = df_categ['Barratt_Barratt_P2_Edu'].fillna(df_categ['Barratt_Barratt_P2_Edu'].mode()[0])  
```

**Splitting our dataset into 70% train and 30% test then doing standard scaling to our training dataset so we make sure all the values are scaled and are in quite close range to prevent the dominance of any feature over the others**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, multilabel_confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from imblearn.over_sampling import SMOTE

# Your original train/test split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42, stratify=Y)

#scaling the training dataset after splitting
scaler = StandardScaler()
#scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

**KNN Model**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, multilabel_confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define base KNN
knn = KNeighborsClassifier()

# Wrap in MultiOutputClassifier
multi_knn = MultiOutputClassifier(knn)

# Define parameter grid
param_grid = {
    'estimator__n_neighbors': [3, 5, 7, 9, 11],
    'estimator__weights': ['uniform', 'distance'],
    'estimator__p': [1, 2]  # p=1: Manhattan, p=2: Euclidean
}

# Run grid search
grid_search = GridSearchCV(multi_knn, param_grid, cv=3, scoring='f1_macro', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Best model
best_model = grid_search.best_estimator_

# Print best parameters
print("Best Parameters Found by GridSearch:")
print(grid_search.best_params_)

# Predict on test set
y_pred = best_model.predict(X_test)

# Get unique class labels for each target (assuming y_test is DataFrame)
unique_classes = [np.unique(y_test.iloc[:, i]) for i in range(y_test.shape[1])]

# Generate separate reports and plots for each target
for i, target in enumerate(Y.columns):
    print(f"\n{'='*50}")
    print(f"Performance Report for Target: {target}")
    print(f"{'='*50}")
    
    # Classification report for current target (shows actual class labels)
    print("\nClassification Report:")
    print(classification_report(y_test.iloc[:, i], y_pred[:, i], 
                              target_names=[str(cls) for cls in unique_classes[i]]))
    
    # Accuracy for current target
    acc = accuracy_score(y_test.iloc[:, i], y_pred[:, i])
    print(f"Accuracy: {acc:.4f}")
    
    # Confusion matrix for current target (with 0/1 labels)
    cm = multilabel_confusion_matrix(y_test, y_pred)[i]
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Predicted 0', 'Predicted 1'],
                yticklabels=['Actual 0', 'Actual 1'])
    plt.title(f'Confusion Matrix: {target}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    plt.show()
```

**Logistic Regression model**

**As we used L1 regularization first to remove unimportant features by forcing some coefficients to zero, simplifying the model. Once these weak features are filtered out i used the L2 regularization which doesnot eliminate more features but smooths the remaining coefficients making the model more stable and accurate. This two-step approach (L1 for feature selection, L2 for final tuning) keeps the model simple while improving performance avoiding the risk of over-sparsifying with L1 twice. To simplify L1 cuts the noise, L2 refines the signal**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, multilabel_confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Initialize the Logistic Regression model
log_reg = LogisticRegression(max_iter=1000, random_state=42)

# Wrap in MultiOutputClassifier for multi-label classification
model = MultiOutputClassifier(log_reg, n_jobs=-1)

# Define hyperparameters for Grid Search
param_grid = {
    'estimator__C': [0.01, 0.1, 1, 10, 100],
    'estimator__penalty': ['l2'],  # Prefer L2 after L1 feature selection
    'estimator__solver': ['liblinear', 'lbfgs']  # 'lbfgs' for L2
}

# Perform Grid Search with cross-validation
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)

# Best model after Grid Search
best_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)

# Predict labels on the test set
y_pred = best_model.predict(X_test)

# Get unique class labels for each target
unique_classes = [np.unique(y_test.iloc[:, i]) for i in range(y_test.shape[1])]

# Generate separate reports and plots for each target
for i, target in enumerate(Y.columns):
    print(f"\n{'='*50}")
    print(f"Performance Report for Target: {target}")
    print(f"{'='*50}")
    
    # Classification report with actual class labels
    print("\nClassification Report:")
    print(classification_report(y_test.iloc[:, i], y_pred[:, i], 
                              target_names=[str(cls) for cls in unique_classes[i]]))
    
    # Accuracy score
    acc = accuracy_score(y_test.iloc[:, i], y_pred[:, i])
    print(f"Accuracy: {acc:.4f}")
    
    # Confusion matrix with 0/1 labels
    cm = multilabel_confusion_matrix(y_test, y_pred)[i]
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Predicted 0', 'Predicted 1'],
                yticklabels=['Actual 0', 'Actual 1'])
    plt.title(f'Confusion Matrix: {target}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    plt.show()
```
